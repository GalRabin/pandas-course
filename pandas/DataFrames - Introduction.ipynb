## Drop null raws

```python
nba = pd.read_csv("nba.csv")
nba.tail()
```

```python
print("Will remove any row with NaN value")
nba.dropna()
```

```python
print("Will remove any row with any NaN value")
nba.dropna(how="any")
```

```python
print("Will remove any row with all NaN value")
nba.dropna(how="all")
```

```python
print("Will remove any row with college NaN value")
nba.dropna(subset=["College"])
```

## Sort values

```python
nba.sort_values(by="Name", ascending=False).head()
```

## Rank

```python
nba = pd.read_csv("nba.csv").dropna(how=="all")
nba.head()
```

```python
nba["Salary"] = nba["Salary"].fillna(0).astype(int)
```

```python
nba["Salary"].rank(ascending=True)
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype(int)
```

```python
nba.sort_values("Salary", ascending=False)
```
